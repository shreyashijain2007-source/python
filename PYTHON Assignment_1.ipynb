Q2. Check Data Types using type()?

    INPUT-
    name = "YOUR NAME"
    number = 100
    decimal = 10.5
    status = False

    OUTPUT-
    <class 'str'>
    <class 'int'>
    <class 'float'>
    <class 'bool'>

```python
name = "dodo"
number = 100
decimal = 10.5
status = False

print(type(name))
print(type(number))
print(type(decimal))
print(type(status))
```

Q4. Integer Type Conversion 

    INPUT-
    num = "25"

    OUTPUT-
    25
    <class 'int'>

```python
num = "25"
num = int(num)

print(num)
print(type(num))
```

Q5. Float Type Conversion

    INPUT-
    num = "10.5"

    OUTPUT-
    10.5
    <class 'float'>

```python
num = "10.5"
num = float(num)

print(num)
print(type(num))
```

Q6. Convert int to float
